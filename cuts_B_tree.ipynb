```python
import sys
import math

def upgrade(arr, i, new):
    i -= 1
    N = len(arr) // 2
    i += N
    arr[i] = new
    i = i // 2
    while i:
        arr[i] = arr[2 * i] + arr[2 * i + 1]
        i = i // 2

def cuts_tree(file):
    temp = file.split('.')[0]
    with open(file, 'r') as f_in:
        with open(temp + '.out', 'w') as f_out:
            line = f_in.readline()
            while line:
                n, k = line.strip().split()
                n = int(n)
                k = int(k)
                N= 2 ** (math.ceil(math.log2(n)))
                arr = [0] * 2 * N
                for _ in range(k):
                    line = f_in.readline().strip().split()
                    if line[0] == 'A':
                        upgrade(arr, int(line[1]), int(line[2]))
#                         print(arr)
                    if line[0] == 'Q':
                        L = int(line[1]) + N - 1
                        R = int(line[2]) + N - 1
                        summ = 0
                        while (L <= R):
                            if (L % 2) != 0:
                                summ += arr[L]
                            L = (L + 1) // 2
                            if (R % 2) == 0:
                                summ += arr[R]
                            R = (R - 1) // 2
                        f_out.write(str(summ) + '\n')
                line = f_in.readline()

# if __name__ == "__main__":
#     cuts_tree(sys.argv[1])

file ='sum.in'
cuts_tree(file)

with open('sum.out', 'r') as f:
    print(f.read())
```
